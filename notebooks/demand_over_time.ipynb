```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\CUIDAIHACK\output\district_month_demand.csv")

# Convert month column
df["month"] = pd.to_datetime(df["year_month"])

# Aggregate demand per month (all districts)
monthly_demand = df.groupby("year_month")["demand_total"].sum()

# Plot
plt.figure()
plt.plot(monthly_demand.index, monthly_demand.values)
plt.xlabel("Month")
plt.ylabel("Total Demand")
plt.title("Monthly Aadhaar Demand Trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
