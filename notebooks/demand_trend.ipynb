```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\CUIDAIHACK\output\district_monthly_trend.csv")
df["month"] = pd.to_datetime(df["year_month"])

plt.figure()
plt.plot(df["month"], df["mom_growth"])
plt.axhline(0)  # baseline
plt.xlabel("Month")
plt.ylabel("MoM Growth (%)")
plt.title("Month-over-Month Demand Growth")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
