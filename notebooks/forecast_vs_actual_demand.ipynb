```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\CUIDAIHACK\output\final_district_month_predictions.csv")
df["month"] = pd.to_datetime(df["month"])

# Aggregate
actual = df.groupby("month")["monthly_total_load"].sum()
predicted = df.groupby("month")["district_3M_forecast"].sum()

plt.figure()
plt.plot(actual.index, actual.values, label="Actual Demand")
plt.plot(predicted.index, predicted.values, label="Predicted Demand")
plt.xlabel("Month")
plt.ylabel("Demand")
plt.title("Actual vs Forecasted Demand")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
