```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\CUIDAIHACK\output\district_spike_alerts.csv")

spike_counts = df["district"].value_counts()

plt.figure()
plt.bar(spike_counts.index, spike_counts.values)
plt.xlabel("District")
plt.ylabel("Spike Count")
plt.title("Demand Spike Alerts by District")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
